```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
test=pd.read_csv('/kaggle/input/mobile-analytics/test_VsU9xXK.csv')
```

```python
test.head()
```

```python
test.drop(['Trip_ID', 'Var1'], 1, inplace=True)
```

```python
test['Gender']=test['Gender'].astype('category')
test['Confidence_Life_Style_Index']=test['Confidence_Life_Style_Index'].astype('category')
test['Destination_Type']=test['Destination_Type'].astype('category')
test['Type_of_Cab']=test['Type_of_Cab'].astype('category')
```

```python
cat_columns = test.select_dtypes(['category']).columns
test[cat_columns] = test[cat_columns].apply(lambda x: x.cat.codes)
```

```python
test.replace(-1, np.nan, inplace=True)
```

```python
test.head(10)
```

```python
test.dtypes
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5, weights='uniform', metric='nan_euclidean')
```

```python
imputer.fit(test)
```

```python
test = imputer.transform(test)
```

```python
test=pd.DataFrame(test)
test.head(10)
```

```python
test.columns=['Trip_Distance', 'Type_of_Cab', 'Customer_Since_Months',
       'Life_Style_Index', 'Confidence_Life_Style_Index', 'Destination_Type',
       'Customer_Rating', 'Cancellation_Last_1Month', 'Var2', 'Var3',
       'Gender']
```

```python
test.head(10)
```

```python
test.to_csv('/kaggle/working/cleaned_test.csv', index=False)
```
